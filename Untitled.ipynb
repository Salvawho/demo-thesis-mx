```python
df2 = pd.read_csv("dataset_PF.csv")
```

```python
df2.head()
```

```python
media_lat = df2['Latitud'].mean()
print(media_lat)
```

```python
df2['Latitud'].fillna(media_lat, inplace = True)
```

```python
media_lon = df2['Longitud'].mean()
df2['Longitud'].fillna(media_lon, inplace=True)
```

```python
df2.head()
```

```python
df2.rename(columns={'Latitud': 'Lat', 'Longitud':'Lon', 'Fecha_Incidente': 'Date/Time'}, inplace=True)
df2.head()
```

```python
df2 = df2.drop(columns=['Fecha_Publicacion', 'Distrito_Policiaco', 'Direccion', 'Unnamed: 17'])
df2
```

```python
df2 = df2.drop(columns=df2.columns[7::])
```

```python
df2.head()
```

```python
df2 = df2.drop(index=df2.index[509::])
```

```python
df2.tail()
```

```python
df2['Date/Time'] = df2['Date/Time'].fillna(df2['Date/Time'].shift())
df2.head()
```

```python
df2['Date/Time'] = pd.to_datetime(df2['Date/Time'] + ' ' + df2['Hora'])
df2.head()
```

```python
df2 = df2.drop(columns=['Categoria'])
df2.head()
```
